```python
# Create features variable 
X =df[['Rooms', 'Distance', 'Bathroom', 'Car', 'Landsize', 
            'BuildingArea', 'Propertycount','Age']]

# Create target variable
y = df['Price']
```

```python
# Train, test, split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = .20, random_state= 0)
```

```python
# Fit
# Import model
from sklearn.linear_model import LinearRegression

# Create linear regression object
regressor = LinearRegression()

# Fit model to training data
regressor.fit(X_train,y_train)
```

```python
# Predict
# Predicting test set results
y_pred = regressor.predict(X_test)
```

```python
accuracy = regressor.score(X_test, y_test)
print(accuracy*100,'%')
```

```python
#Regression Evaluation Metrics

#Mean Absolute Error (MAE)
#Mean Squared Error (MSE)
#Root Mean Squared Error (RMSE)
```

```python
from sklearn import metrics
print('MAE:',metrics.mean_absolute_error(y_test,y_pred))
print('MSE:',metrics.mean_squared_error(y_test,y_pred))
print('RMSE:',np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```

```python
# Calculated R Squared
print('R^2 =',metrics.explained_variance_score(y_test,y_pred))
```

```python
# Actual v predictions scatter
plt.scatter(y_test, y_pred)
```

```python
# Histogram of the distribution of residuals
sns.distplot((y_test - y_pred))
```

```python
y = df.Price;
predictors = ['Rooms', 'Bathroom', 'Landsize', 'BuildingArea', 
                        'YearBuilt', 'Lattitude', 'Longtitude']
x = df[predictors]
```

```python
#The below step replaces NAN values with median
X = x.iloc[:, :].values 
from sklearn.preprocessing import Imputer
imputer = Imputer(missing_values='NaN',strategy='median',axis=0)
imputer = imputer.fit(X[:, :])
X[:, :] = imputer.transform(X[:, :])
```

```python
#Fit model
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error
model = DecisionTreeRegressor()
model.fit(X, y)
xHead = x.head()
XHead = xHead.iloc[:,:].values
imputer = Imputer(missing_values='NaN',strategy='median',axis=0)
imputer = imputer.fit(XHead[:, :])
XHead = imputer.transform(XHead[:,:])
predicted_price = model.predict(X)
b_mae=mean_absolute_error(y, predicted_price)
print(b_mae)#Print mean absolute error in predicting the price of a house
from sklearn.model_selection import train_test_split
```

```python
#Calculating mean absolute error for various training data
train_X, val_X, train_y, val_y = train_test_split(X, y,random_state = 0)
model.fit(train_X, train_y)
t_predictions = model.predict(val_X)
print(mean_absolute_error(val_y, t_predictions))

#Function to see how many max leaf nodes in Decision tree and their mean absolute error
def get_mae(max_leaf_nodes, predictors_train, predictors_val, targ_train, targ_val):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(predictors_train, targ_train)
    preds_val = model.predict(predictors_val)
    mae = mean_absolute_error(targ_val, preds_val)
    return(mae)
```

```python
#Calculating the mean absolute error for max leaf nodes 5, 50 500 and 5000
for max_leaf_nodes in [5, 50, 500, 5000]:
    my_mae = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print("Max leaf nodes: %d  \t\t Mean Absolute Error:  %d" %(max_leaf_nodes, my_mae))
```
